```python
# ims = io.imread('data/branch/OakBranch.tif')
ims = io.imread("data/coral/MexCoral.tif")
```

```python
gt = ims[300,:,:]

w, h = gt.shape
print(f"GT shape: {gt.shape}")

lr = resize(gt, (w // 4, h // 4))
print(f"LR shape: {lr.shape}")
```

```python
io.imshow(gt, cmap="gray")
io.show()
io.imshow(lr, cmap="gray")
io.show()
```

```python
io.imsave("results/gt.png", gt.astype("uint8"))
io.imsave("results/lr.png", lr.astype("uint8"))
```
